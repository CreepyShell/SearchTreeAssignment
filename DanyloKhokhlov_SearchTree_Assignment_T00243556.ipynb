## CurrentBoard class

```python
import copy
class Board:
  def __init__(self, board = None):
    #this value means that a cell in a board in empty and be be used by a player to make a move
    self.default = -1
    self.width = 7
    self.height = 6
    if board is not None:
      self.board = board
    else:
      self.board = [[ self.default for j in range(0,self.width)] for i in range(0,self.height)]

  #function to display current board. With a board it shows all possible moves that can be done by a player
  def display_board(self):
    print("Possible moves: " + ", ".join([str(move) for move in self.get_all_possible_moves()]))
    for i in range(0, self.height):
      print(' | '.join([f" {self.board[i][j]}" if self.board[i][j] != self.default else str(self.board[i][j]) for j in range(self.width)]))
      print("\n")

  #function to update the board array after a player made a turn
  def play(self, player, column_num):
    #checking if there is an empty cell where player wants to move
    if self.board[0][column_num - 1] != self.default:
      return False

    #checking if the bottom of a board is empty
    if self.board[self.height - 1][column_num - 1] == self.default:
      #assign a player to the bottom of a board
      self.board[self.height - 1][column_num - 1] = player
      return True

    #iterate thought the available cell until an opponent value is found
    for index in range(1, self.height):
      if self.board[index][column_num - 1] != self.default:
        #assign current player value upper the opponent value
        self.board[index - 1][column_num - 1] = player
        return True

  def get_opponent(self, player):
    if player == 1:
      return 0
    return 1

  def get_all_possible_moves(self):
    moves = []
    #iterate thought the top of a board to get all empty cells (where a player can make a move)
    for i in range(self.width):
      if self.board[0][i] == self.default:
        moves.append(i + 1)
    return moves

  #function to check if a player has win in one move. Basically it makes all possible move for a player and checks
  #if a game is already won by a player. It returns the win move for a player
  def is_one_move_winning(self, player):
    for move in self.get_all_possible_moves():
      board = copy.deepcopy(self)
      board.play(player, move)
      if board.check_win(player):
        return move
    return -1

  #function to check if a player is won. Returns true if they win, false if not
  def check_win(self, player):
    #checking horisontal win
    for row in range(self.height):
      hor_count = 0
      for column in range(self.width):
          if self.board[row][column] == player:
              hor_count += 1
          else:
              hor_count = 0
          if hor_count == 4:
              return True

    #checking vertical win
    for column in range(self.width):
      vert_count = 0
      for row in range(self.height):
          if self.board[row][column] == player:
              vert_count += 1
          else:
              vert_count = 0
          if vert_count == 4:
              return True

    #checking diagonal win
    row = 0
    column = 0
    cross_count = 0
    for i in range((self.width - 3) * (self.height - 3) * 4):
      if self.board[row][column] == player:
        cross_count += 1
      else:
        cross_count = 0
      if cross_count == 4:
        return True
      row += 1
      column += 1
      #move to the next diagonal
      if (i + 1) % 4 == 0:
        row -= 4
        column -= 3
        cross_count = 0
      #move to the next diagonal and row
      if (i + 1) % 16 == 0:
        row = int((i + 1) / 16)
        column = 0

    #checking the reverse diagonal win
    row = 0
    column = self.width - 1
    revese_cross_count = 0
    for i in range((self.width - 3) * (self.height - 3) * 4):
      if self.board[row][column] == player:
        revese_cross_count += 1
      else:
        revese_cross_count = 0
      if revese_cross_count == 4:
        return True
      row += 1
      column -= 1
      #move to the next diagonal
      if (i + 1) % 4 == 0:
        row -= 4
        column += 3
        revese_cross_count = 0
      #move to the next diagonal and row
      if (i + 1) % 16 == 0:
        row = int((i + 1) / 16)
        column = self.width - 1

    return False
```

## Search Tree class

```python
class SearchTreeNode:
  def __init__(self, board, player, depth = 0):
    #value for maximum depth, up to 7 depth was tested (took 3 minutes to generate a tree with all nodes)
    self.max_depth = 6

    #children for each tree node. if a tree is a won board, this tree has no children
    self.children = []

    self.current_board = board
    self.player = player

    #the depth of current tree node
    self.depth = depth
    self.value = None

    #win move to an already won board
    self.win_move = self.current_board.is_one_move_winning(self.player)

    #checking if a currect node has a won board
    if self.win_move != -1:
      #assigning a value to a tree node depending on a depth - first move is always by a computer (depth % 2 == 0), the second by a player (depth % 2 == 1)
      if self.depth % 2 == 0:
        self.value = 1
      else:
        self.value = -1
    else:
      #if a tree reaches maximum depth, stop generating children
      if self.depth < self.max_depth:
        self.generate_children()
      else:
        #
        self.value = 0

  def generate_children(self):
    #find all possible board states after a current player made a move
    for move in self.current_board.get_all_possible_moves():
      opponent = self.current_board.get_opponent(self.player)
      child_board = copy.deepcopy(self.current_board)
      child_board.play(self.player, move)
      self.children.append(SearchTreeNode(child_board, opponent, self.depth + 1))

  #function for finding the best possible move by a computer
  def sort_values(self):
    #if a tree node is the last in all tree or a value is already assigned, just return it
    if self.value is not None:
      return self.value

    #sorting values from the worst to the best
    self.children = sorted(self.children, key = lambda x: x.sort_values())

    #assinging the best possible value to the tree node that the most profitable for a computer
    if self.depth % 2 == 0:
      #for a computer move the tree value has to be the greatest of all children
      self.value = self.children[-1].value
    else:
      #The explanation of this evaluation function is written in the conclusion section
      neg_counts = sum([child.value for child in self.children if child.value < 0])
      pos_counts = sum([child.value for child in self.children if child.value > 0])
      self.value = (neg_counts + pos_counts) / len(self.children)
    return self.value

  #function to get the next move for a computer. Because a tree stops generating children
  #after a board has win in one move this win move has to be made
  def get_next_move(self):
    if len(self.children) == 0:
      self.current_board.play(self.player, self.win_move)
      return self.current_board
    return self.children[-1].current_board
```

##  Gameplay

```python
def play_connect4():
  response = input("Do you wish to play first (y/n) ?")
  players_turn  = (response == "y")

  response = input("Do you wish to play 1 or 0  (1/0) ?")
  cb = Board()
  cb.display_board()
  player = int(response)
  computer = cb.get_opponent(player)

  finish = False
  #game can be finished by win one of the players, draw or if a person chooses a move outside possible range
  while (not finish):
    if players_turn:
      choice = input("Make your move ")
      if int(choice) not in cb.get_all_possible_moves():
        print("ok done")
        break
      cb.play(player, int(choice))
    else:
      search_tree = SearchTreeNode(cb, computer, 0)
      search_tree.sort_values()
      cb = search_tree.get_next_move()
      cb.display_board()
    if cb.check_win(computer):
      print("You lost")
      finish = True
    if cb.check_win(player):
      print("You won")
      finish = True
    if len(cb.get_all_possible_moves()) == 0:
      print("draw")
      finish = True
    players_turn = not players_turn
```

<h4>Search Tree Nodes Evaluation Explanation</h4>

Each node in the search tree must have an assigned value to reflect the strength of the current board state. However, once the maximum depth is reached, accurately evaluating the board becomes challenging. At that point, positions may involve complex patterns that are hard to detect and interpret programmatically. Instead of implementing a detailed and time-consuming evaluation function, a simpler approach was chosen: calculate the overall percentage of outcomes in which either the computer or the player wins. These values can be positive or negative, and if the number of wins is roughly equal for both sides, the node's value tends toward zero. The computer then chooses the node with the highest average score, preferring paths that most often lead to a win or, at the very least, a draw.

For example, if a node has children whose descendants mostly result in the player winning, that node will receive a negative average score—signaling it as unfavorable. Conversely, if most of its descendants are winning positions for the computer, the score will be positive, making it an attractive option. This guides the algorithm to avoid branches where the player tends to win and focus on those offering higher chances of victory for the computer.

To keep it simple:
- It avoids harsh decisions based on one bad outcome (unlike strict minimax).
- It captures the overall favorability of a node.
- If the player wins in many subnodes, the average drops (signaling danger).
- If the computer wins often, the average rises (indicating opportunity).

Rather than reacting to a single worst-case scenario, this evaluation balances risk and reward. It selects moves based on how frequently they lead to success, offering a more stable and realistic decision-making process.

<h3>Algorithm running and perfomance evaluating</h3>

In two test runs, the algorithm successfully led the computer to victory, demonstrating that it performs well in practice. However, this does not necessarily imply that the algorithm is flawless. While the outcomes suggest a strong decision-making process, the current evaluation strategy can overlook critical tactical positions that guarantee a win if followed precisely. Much like in chess, where achieving checkmate might involve sacrificing key pieces such as the queen or rooks, a winning strategy in Connect 4 may require short-term risks or seemingly suboptimal moves. If the algorithm doesn't recognize these sequences, it may choose a safer but ultimately weaker path.

A key limitation of the current evaluation method is that it averages the values of child nodes, which can cause the algorithm to avoid certain winning nodes if those nodes also contain many branches where the player wins. As a result, it may skip over a guaranteed win because the surrounding statistics suggest risk. To overcome this, the evaluation function could be enhanced to include more detailed board analysis—such as recognizing potential alignments, counting consecutive tokens, or assessing open spaces near the computer’s cells. Additionally, assigning a higher score to winning paths that are a few moves deep (rather than immediate wins) could encourage the algorithm to pursue longer-term victories more confidently
